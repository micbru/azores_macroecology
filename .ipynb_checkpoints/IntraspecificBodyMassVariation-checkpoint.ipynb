This file looks at the body length variances in the beetle and spider datasets. The goal is to get an idea of what these distributions look like, so we can model the variance to get appropriate MRDIs to compare to METE.

```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beetles = pd.read_csv('./RawData/beetles_body_size_indivs.csv')
spiders = pd.read_csv('./RawData/spiders_body_size_indivs.csv')
```

# Beetles

### Individual species

```python
# Now for individual species, which is what we actually want
bspm = beetles.groupby('Species')['Body_mass']
# means
bm = bspm.mean()
bv = bspm.var()
```

# Spiders

### Individual species

```python
# Now for individual species, which is what we actually want
sspm = spiders.groupby('Species')['Body_mass']
# means
sm = sspm.mean()
sv = sspm.var()
```

# Plot for SI
Show the four most abundant beetles and spiders and mean variance relationship

```python
# Mean variance on one plot

fig,ax = plt.subplots(1,2,constrained_layout=True,sharey=True,sharex=True,figsize=(8,3))

# Beetles
ax[0].set_title('Coleoptera')
ax[0].plot(np.log10(bm),np.log10(bv),'o')
ax[0].set_xlabel(r'$\log_{10}$(Mean)')
ax[0].set_ylabel(r'$\log_{10}$(Variance)')
# Regression
beetle_fit = st.linregress(np.log10(bm.values),np.log10(bv.values))
bf = pd.Series(beetle_fit,index=['Slope','Intercept','Rvalue','pvalues','Stderr'])
xplot = np.linspace(np.log10(min(bm)),np.log10(max(bm)),100)
ax[0].plot(xplot, beetle_fit.intercept + beetle_fit.slope*xplot)

# Spiders
ax[1].set_title('Araneae')
ax[1].plot(np.log10(sm),np.log10(sv),'o')
ax[1].set_xlabel(r'$\log_{10}$(Mean)')
ax[1].set_ylabel('')
# Regression
spider_fit = st.linregress(np.log10(sm.values),np.log10(sv.values))
spf = pd.Series(spider_fit,index=['Slope','Intercept','Rvalue','pvalues','Stderr'])
xplot = np.linspace(np.log10(min(sm)),np.log10(max(sm)),100)
ax[1].plot(xplot, spider_fit.intercept + spider_fit.slope*xplot)

fig.savefig('Figures/SI/VarianceMean.pdf')
```

```python
# Print out results to use in reconstructing MRDIs
print("Beetle")
print(bf)
print()
print("Spider")
print(spf)
```

```python
# Plot below
# nmost abundant
nm = 4
# get the most abundance
bmost_ab1 = bspm.count().sort_values()[-nm::].index
smost_ab1 = sspm.count().sort_values()[-nm::].index

# Make figure
fig,axs = plt.subplots(2,nm,constrained_layout=True,figsize=(10,6))

# Plot beetles
for s,ax in zip(bmost_ab1,axs.flatten()[0:nm]):
    beetles.loc[(beetles['Species']==s),'Body_mass'].plot(kind='hist',density=True,ax=ax,label='Hist')
#    beetles.loc[(beetles['Species']==s),'Body_mass'].plot(kind='kde',ax=ax,label='KDE')
    ax.set_title(s)
    ax.set_xlabel('Body mass (mg)')
    ax.set_ylabel('')
    # What if I just pretend these are normal?
    m = bm.loc[s]
    std = np.sqrt(bv.loc[s])
    x = np.linspace(m-3*std,m+3*std,100)
    ax.plot(x,st.norm.pdf(x,m,std),label='Normal')
# Legend and labels
axs[0,nm-1].legend()
axs[0,0].set_ylabel('Probability density')
axs[0,0].annotate('Coleoptera', xy=(0, 0.5), xytext=(-axs[0,0].yaxis.labelpad - 5, 0),
                xycoords=axs[0,0].yaxis.label, textcoords='offset points',
                size='large', ha='right', va='center')

# Plot spiders
for s,ax in zip(smost_ab1,axs.flatten()[nm::]):
    spiders.loc[(spiders['Species']==s),'Body_mass'].plot(kind='hist',density=True,ax=ax,label='Hist')
#    spiders.loc[(spiders['Species']==s),'Body_mass'].plot(kind='kde',ax=ax,label='KDE')
    ax.set_title(s)
    ax.set_xlabel('Body mass (mg)')
    ax.set_ylabel('')
    # What if I just pretend these are normal?
    m = sm.loc[s]
    std = np.sqrt(sv.loc[s])
    x = np.linspace(m-3*std,m+3*std,100)
    ax.plot(x,st.norm.pdf(x,m,std),label='Normal')
# legends nad labels
#axs[1,nm -1].legend()
axs[1,0].set_ylabel('Probability density')
axs[1,0].annotate('Araneae', xy=(0, 0.5), xytext=(-axs[1,0].yaxis.labelpad - 5, 0),
                xycoords=axs[1,0].yaxis.label, textcoords='offset points',
                size='large', ha='right', va='center')

fig.savefig("Figures/SI/Most_Abundant.pdf")
```
