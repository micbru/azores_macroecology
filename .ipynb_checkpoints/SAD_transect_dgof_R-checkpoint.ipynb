Since we really should use dgof for the SADs due to the repeated values, let's do that here for all transects.

```r
# Load in library for SADs and dgof
library(sads)
library(extraDistr)
library(dgof)
```

```r
sn <- read.csv('ProcessedData/sad_data.csv')
abd <- read.csv('ProcessedData/sad_abd.csv',skip=1)
# Now remove first row, which just tells you X is MF (pandas thing)
abd = abd[-1,]
# For some reason these are with dots rather than dashes, but we can deal with that
```

```r
sn
```

```r
# Make KS arrays for each land use
ks_ef <- c()
ks_nf <- c()
ks_ip <- c()
ks_sp <- c()
# Loop over rows
for (i in 1:nrow(sn)){
    site = sn[i,'X.1']
    # Substitute - with . so we can read properly
    st = gsub("-",".",site)
    # Get temporary abundance and only non-zero values
    abdt = abd[st]
    abdt = abdt[abdt!=0]
    # Get corresponding beta
    bt = sn[i,'beta']
    # Now do KS test
    kst <- dgof::ks.test(abdt,stepfun(1:max(abdt),plgser(0:max(abdt),exp(-bt))))
    # Get just the D statistic
    ksD <- unname(unlist(kst['statistic']))
    # or 
    #    stats::ks.test(abd2,stepfun(1:max(abd2),plgser(0:max(abd2),exp(-beta))))
    # Now save to appropriate list based on land use
    if (sn[i,'X']=="Exotic forest")
        ks_ef <- c(ks_ef,ksD)
    if (sn[i,'X']=="Native forest")
        ks_nf <- c(ks_nf,ksD)
    if (sn[i,'X']=="Intensive pasture")
        ks_ip <- c(ks_ip,ksD)
    if (sn[i,'X']=="Semi-natural pasture")
        ks_sp <- c(ks_sp,ksD)
}
```

```r
# Show a boxplot
boxplot(ks_nf,ks_ef,ks_sp,ks_ip,names=c('Native forest','Exotic forest','Semi pasture','Intensive pasture'),
       ylab='KS D')
```

```r
# Get means and standard errors and save to file
# Do not plot here since it's actually pretty annoying in R and I don't want to learn ggplot right now.
means <- c(mean(ks_nf),mean(ks_ef),mean(ks_sp),mean(ks_ip))
se <- c(sd(ks_nf)/sqrt(length(ks_nf)),sd(ks_ef)/sqrt(length(ks_ef)),
        sd(ks_sp)/sqrt(length(ks_sp)),sd(ks_ip)/sqrt(length(ks_ip)))
df <- data.frame(means,se)
rownames(df) = c('Native forest','Exotic forest','Semi-natural pasture','Intensive pasture')
print(df)
write.csv(df,"ProcessedData/sad_means_ksD_R.csv", row.names = TRUE)
```
